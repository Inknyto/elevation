Create the entreprises index

```python
# Elasticsearch server details

import requests
from my_credentials import *
base_url = "http://localhost:9200"
index_name = "senegal_entreprises_data"
```

```python


# Index settings and mapping

index_settings = {
    "settings": {
        "number_of_shards": 1,
        "number_of_replicas": 0
    },
    
    "mappings": {
        
        "properties": {

            'Nom': {"type": "keyword"},
            'Type': {"type": "keyword"},
            'Region': {"type": "keyword"},
            'Arrondissement': {"type": "keyword"},
            'Commune': {"type": "keyword"},
            'Quartier': {"type": "keyword"},
            'TÃ©l': {"type": "keyword"},
            'Statut ligne': {"type": "keyword"}
            
        }
    }
}

# Create the index with authentication
response = requests.put(
    f"{base_url}/{index_name}",
    json=index_settings,
    auth=(username, password)  # Include authentication parameters
)

# Check the response
if response.status_code == 200:
    print(f"Index '{index_name}' created successfully.")
else:
    print(f"Error creating index: {response.content}")
```

Post the entreprise data to the index

```python
import json
elasticsearch_url = "http://localhost:9200"
with open('entreprise_data_2.json') as f:
    cleaned = json.load(f)
for c in cleaned:
        response = requests.post(
            f"{elasticsearch_url}/senegal_entreprises_data/_doc",
            json=c,
            auth=(username, password)  # Include authentication parameters

        )

        if response.status_code == 201:
            print(f"Entreprise data uploaded successfully: {response.json()}")
        else:
            print(f"Error uploading entreprise data: {response.json()}")
```

Retrieve all data from the index

```python
# Define the search query to match all documents
query = {
    "query": {
        "match_all": {}
    }
}

# Send the search request
response = requests.get(
    f"{base_url}/{index_name}/_search",
    json=query,
    auth=(username, password)  # Include authentication parameters
)

# Check the response
if response.status_code == 200:
    data = response.json()
    hits = data.get("hits", {}).get("hits", [])

    if hits:
        for hit in hits:
            print(f"Document ID: {hit['_id']}")
            print(f"Document Content: {hit['_source']}")
    else:
        print("No documents found in the index.")
else:
    print(f"Error fetching data: {response.content}")
```
